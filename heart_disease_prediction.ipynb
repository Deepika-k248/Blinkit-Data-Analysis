data collection

```python
df=pd.read_csv('Heart_Disease_Prediction.csv')
print (df)
```

data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['Sex']=le.fit_transform(df['Sex'])
df['Chest pain type']=le.fit_transform(df['Chest pain type'])
df['Exercise angina']=le.fit_transform(df['Exercise angina'])
# Correcting the column names to match the DataFrame
df['Slope of ST']=le.fit_transform(df['Slope of ST'])
df['EKG results']=le.fit_transform(df['EKG results'])
print (df)
```

```python
x = df.drop('Heart Disease', axis=1)
y = df['Heart Disease']
print("xxxx",x)
```

```python
print("yyyy",x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=12)
print("df",df.shape)
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

model training

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

model evaluation

```python
y_pred=nb.predict(x_test)
print("y_pred",y_pred)
print("y_test",y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
print("accuracy",accuracy)
```

model prediction

```python
import numpy as np

# The model expects 13 features. The input array had 11 features.
# Adding placeholder values for 'Number of vessels fluro' and 'Thallium'
test_prediction = nb.predict(np.array([29,0,2,100,106,1,2,80,1,1,1, 0, 0]).reshape(1, -1))
if test_prediction[0] == 'Presence':
  print("the patient have heart disease,please consult the doctor")
else:
  print("the patient is normal")
```
